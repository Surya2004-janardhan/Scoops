```python
from groq import Groq

client = Groq()

print(previous_msgs)
chat_completion = client.chat.completions.create(
    messages=[
        # Set an optional system message. This sets the behavior of the
        # assistant and can be used to provide specific instructions for
        # how it should behave throughout the conversation.
        {
            "role": "system",
            "content": "You are a helpful assistant! max words 30"
        },
        # Set a user message for the assistant to respond to.
        {
            "role": "user",
            # along with query previou msgs inlcude 
            "content":  f"Context : {previous_msgs}   query : wts my study "
        }
    ],

    # The language model which will generate the completion.
    model="llama-3.3-70b-versatile"
)
# print(chat_completion)
# Print the completion returned by the LLM.


previous_msgs.append(chat_completion.choices[0].message.content)
[   3, 4, 5, 6, 7   ]
if len(previous_msgs) > 5:
    previous_msgs.pop(0)
print(chat_completion.choices[0].message.content)
```
