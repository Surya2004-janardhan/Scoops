# 🚀 Production-Grade RAG (Retrieval-Augmented Generation) Application

## 📚 What is RAG?

**RAG (Retrieval-Augmented Generation)** is an AI framework that enhances Large Language Models (LLMs) by combining them with external knowledge retrieval systems.

### 🔍 How RAG Works:
1. **Ingestion Phase**: Raw documents (PDFs, text files) are processed, chunked, and converted into vector embeddings
2. **Storage Phase**: These embeddings are stored in a vector database for efficient similarity search
3. **Retrieval Phase**: When a user asks a question, the system finds the most relevant document chunks
4. **Generation Phase**: The LLM uses retrieved context to generate accurate, grounded responses

### 🎯 Why RAG?
| Problem with Pure LLMs | RAG Solution |
|------------------------|--------------|
| Knowledge cutoff date | Access to current documents |
| Hallucinations | Grounded responses from real data |
| No domain knowledge | Custom knowledge base integration |
| Generic responses | Context-specific answers |

### 🏗️ Architecture Overview:
```
┌─────────────┐     ┌──────────────┐     ┌─────────────┐
│   PDF Doc   │ ──► │  Text Chunks │ ──► │  Embeddings │
└─────────────┘     └──────────────┘     └─────────────┘
                                                │
                                                ▼
┌─────────────┐     ┌──────────────┐     ┌─────────────┐
│  Response   │ ◄── │   LLM (Groq) │ ◄── │ Vector DB   │
└─────────────┘     └──────────────┘     └─────────────┘
                           ▲
                           │
                    ┌──────────────┐
                    │  User Query  │
                    └──────────────┘
```

# 🧮 Step 3: Embeddings Model Setup

## 🎯 What are Embeddings?

**Embeddings** are dense vector representations of text that capture semantic meaning in a high-dimensional space.

### 📐 How Embeddings Work:

```
Input Text: "The quick brown fox"
                    ↓
          ┌─────────────────┐
          │  Transformer    │
          │  Neural Network │
          └─────────────────┘
                    ↓
Output Vector: [0.23, -0.45, 0.12, ..., 0.89]
               └──────── 384 dimensions ──────┘
```

### 🔬 Technical Details:

**Cosine Similarity** - How we compare vectors:
```
similarity = (A · B) / (||A|| × ||B||)

Range: -1 to 1
  • 1.0  = Identical meaning
  • 0.0  = Unrelated
  • -1.0 = Opposite meaning
```

### 📊 Why `all-MiniLM-L6-v2`?

| Aspect | Details |
|--------|---------|
| **Architecture** | MiniLM (distilled from larger models) |
| **Layers** | 6 transformer layers |
| **Parameters** | ~22 million (lightweight) |
| **Embedding Dimension** | 384 |
| **Max Sequence Length** | 256 tokens |
| **Performance** | Excellent speed/quality tradeoff |
| **Use Case** | Semantic similarity, search, clustering |

### 🏆 Embedding Model Comparison:

| Model | Dimensions | Speed | Quality | Memory |
|-------|------------|-------|---------|--------|
| `all-MiniLM-L6-v2` | 384 | ⚡ Fast | Good | Low |
| `all-mpnet-base-v2` | 768 | Medium | Better | Medium |
| `e5-large-v2` | 1024 | Slow | Best | High |
| `text-embedding-ada-002` | 1536 | API | Excellent | N/A |

**For RAG:** `all-MiniLM-L6-v2` provides the **best balance** of speed and quality for most use cases.

````python
# ========================================
# 🧮 STEP 3: INITIALIZE EMBEDDINGS MODEL
# ========================================

print("🧮 Initializing Embeddings Model...")
print("=" * 50)

# HuggingFaceEmbeddings - Wrapper for sentence-transformers models
# Downloads and loads the model from HuggingFace Hub
# The model converts text to dense vector representations

print("📥 Loading embedding model from HuggingFace Hub...")
print("   • Model: all-MiniLM-L6-v2")
print("   • Source: sentence-transformers")
print("   • Download size: ~90 MB (first time only)")

# Initialize the embedding model
# model_name: Identifier for the pre-trained model on HuggingFace
# This model is specifically trained for semantic similarity tasks
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"  # Lightweight, fast, good quality
)

print("✅ Embedding model loaded successfully!")
print("")

# Test the embedding model with a sample text
print("🧪 Testing embedding generation...")
test_text = "This is a test sentence for embedding"

# embed_query(): Converts a single text string to a vector
# Returns a list of floats representing the text in vector space
test_embedding = embeddings.embed_query(test_text)

print(f"   • Input: '{test_text}'")
print(f"   • Output dimensions: {len(test_embedding)}")
print(f"   • Sample values: [{test_embedding[0]:.4f}, {test_embedding[1]:.4f}, ..., {test_embedding[-1]:.4f}]")
print(f"   • Vector type: {type(test_embedding).__name__}")

print("=" * 50)
print("🎉 Embeddings model ready!")
print(f"📊 Each text chunk will be converted to a {len(test_embedding)}-dimensional vector")
````

## 📖 Understanding PyPDFLoader:

**What it does:**
1. Opens PDF file using `pypdf` library
2. Extracts text from each page
3. Creates `Document` objects with:
   - `page_content`: The extracted text
   - `metadata`: Page number, source file, etc.

**Document Object Structure:**
```python
Document(
    page_content="This is the text from page 1...",
    metadata={
        "source": "document.pdf",
        "page": 0  # 0-indexed
    }
)
```

## ✂️ Understanding RecursiveCharacterTextSplitter:

### 🎯 Why Split Documents?

| Problem | Solution with Splitting |
|---------|------------------------|
| LLMs have token limits | Chunks fit within context window |
| Large docs = slow retrieval | Smaller chunks = faster search |
| Irrelevant info dilutes answers | Precise chunks = focused context |

### 📐 Splitting Parameters:

| Parameter | Description | Our Value |
|-----------|-------------|-----------|
| `chunk_size` | Maximum characters per chunk | 1000 |
| `chunk_overlap` | Characters shared between chunks | 200 |

**Why Overlap?**
```
Without overlap:              With overlap:
┌────────┬────────┐          ┌────────┬────────┐
│ Chunk1 │ Chunk2 │          │ Chunk1 ◄──►Chunk2│
└────────┴────────┘          └────────┴────────┘
Context lost at boundary!     Context preserved!
```

### 🔄 Recursive Splitting Logic:
The splitter tries separators in this order:
1. `"\n\n"` - Double newline (paragraphs)
2. `"\n"` - Single newline
3. `" "` - Space (words)
4. `""` - Individual characters (last resort)

This **preserves semantic boundaries** when possible!

### ✂️ Text Splitting Configuration

**Chunk Size Considerations:**

| Chunk Size | Pros | Cons |
|------------|------|------|
| Small (200-500) | Precise retrieval | May lose context |
| Medium (500-1000) | Good balance | Standard choice |
| Large (1000-2000) | More context | Less precise, more tokens |

**Our Configuration:**
- `chunk_size=1000` - Each chunk is ~1000 characters (~250 words)
- `chunk_overlap=200` - 200 characters shared between adjacent chunks

This ensures we capture enough context while maintaining precision.

````python
# ========================================
# ✂️ STEP 4B: TEXT SPLITTING (CHUNKING)
# ========================================

print("✂️ Starting Text Splitting Process...")
print("=" * 50)

# RecursiveCharacterTextSplitter: Smart text chunking utility
# "Recursive" because it tries multiple separators in order
# Preserves semantic structure by splitting at natural boundaries

print("📐 Configuring Text Splitter:")
print("   • chunk_size: 1000 characters (~250 words)")
print("   • chunk_overlap: 200 characters (20% overlap)")
print("")

# Initialize the text splitter
# chunk_size: Maximum number of characters in each chunk
# chunk_overlap: Number of characters to overlap between chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,      # Max chars per chunk
    chunk_overlap=200     # Overlap between chunks
)

print("✅ Text splitter configured!")
print("")

# split_documents(): Takes a list of Document objects
# Returns: List[Document] - more, smaller Documents
# Metadata is preserved and enhanced with chunk info
print("🔄 Splitting documents into chunks...")
print(f"   📥 Input: {len(documents)} documents")

chunks = text_splitter.split_documents(documents)

print(f"   📤 Output: {len(chunks)} chunks")
print("")

# Calculate and display statistics
print("📊 Chunking Statistics:")
chunk_lengths = [len(chunk.page_content) for chunk in chunks]
print(f"   • Total chunks created: {len(chunks)}")
print(f"   • Average chunk size: {sum(chunk_lengths) // len(chunks)} characters")
print(f"   • Smallest chunk: {min(chunk_lengths)} characters")
print(f"   • Largest chunk: {max(chunk_lengths)} characters")
print("")

# Show sample chunks
print("📝 Sample Chunks Preview:")
for i, chunk in enumerate(chunks[:3]):
    preview = chunk.page_content[:100].replace('\n', ' ')
    print(f"   Chunk {i+1}: '{preview}...'")
    print(f"           (Size: {len(chunk.page_content)} chars, Source: {chunk.metadata.get('source', 'N/A')})")
    print("")

print("=" * 50)
print(f"✂️ Text splitting complete! Created {len(chunks)} chunks for embedding")
````

## 💾 ChromaDB Vector Database

### 🎯 What is ChromaDB?

**ChromaDB** is an open-source, lightweight vector database designed specifically for AI applications. It stores embeddings and enables fast similarity search.

### 🔬 How Vector Search Works:

```
User Query: "What is RAG?"
           ↓
    Query Embedding: [0.12, -0.34, 0.56, ...]
           ↓
    ┌─────────────────────────────┐
    │       ChromaDB              │
    │  ┌─────┐ ┌─────┐ ┌─────┐   │
    │  │Vec 1│ │Vec 2│ │Vec 3│   │  ← All stored embeddings
    │  └─────┘ └─────┘ └─────┘   │
    │     ↑       ↑       ↑      │
    │    0.92   0.45    0.87     │  ← Similarity scores
    └─────────────────────────────┘
           ↓
    Return Vec 1 & Vec 3 (most similar)
```

### 📊 ChromaDB Key Features:

| Feature | Description |
|---------|-------------|
| **In-memory storage** | Fast queries, no external DB needed |
| **Persistent storage** | Optional disk persistence |
| **Metadata filtering** | Filter by source, page, date, etc. |
| **HNSW algorithm** | O(log n) similarity search |
| **No infrastructure** | Runs locally, no Docker required |

### 🔧 Chroma.from_documents() Parameters:

| Parameter | Purpose |
|-----------|---------|
| `documents` | List of Document objects to embed and store |
| `embedding` | The embedding model to use |
| `collection_name` | Name for this vector collection |
| `persist_directory` | (Optional) Directory to save data |

### 📐 Understanding HNSW Algorithm:
**HNSW (Hierarchical Navigable Small World)** is the search algorithm ChromaDB uses:
- Creates a **multi-layer graph** of vectors
- Top layers have **fewer, well-connected nodes** (fast navigation)
- Bottom layers have **all nodes** (accurate search)
- Achieves **O(log n)** search complexity instead of O(n)

````python
# ========================================
# 💾 STEP 4C: CHROMADB VECTOR DATABASE SETUP
# ========================================

print("💾 Initializing ChromaDB Vector Database...")
print("=" * 50)

# Chroma.from_documents(): Factory method to create a vector store
# This single call does multiple things:
# 1. Creates a new ChromaDB collection
# 2. Generates embeddings for all document chunks
# 3. Stores embeddings with their corresponding text and metadata
# 4. Builds the HNSW index for fast similarity search

print("🔄 Starting embedding and storage process...")
print(f"   📊 Documents to process: {len(chunks)} chunks")
print(f"   🧮 Embedding model: all-MiniLM-L6-v2 (384 dimensions)")
print(f"   📁 Collection name: rag_collection")
print("")

# Track the embedding process
import time
start_time = time.time()

print("⏳ Generating embeddings for each chunk...")
print("   (This may take a moment for large documents)")

# Create the vector store
# documents: Our chunked documents
# embedding: The embedding model instance
# collection_name: Identifier for this collection in ChromaDB
vectorstore = Chroma.from_documents(
    documents=chunks,              # The document chunks to embed
    embedding=embeddings,          # Our HuggingFace embedding model
    collection_name="rag_collection"  # Name for the vector collection
)

elapsed_time = time.time() - start_time

print("✅ Vector store created successfully!")
print("")
print("📊 Vector Database Statistics:")
print(f"   • Documents stored: {len(chunks)}")
print(f"   • Embedding dimensions: 384")
print(f"   • Processing time: {elapsed_time:.2f} seconds")
print(f"   • Average time per chunk: {(elapsed_time/len(chunks))*1000:.2f} ms")
print("")

# Create the retriever interface
# as_retriever(): Converts vectorstore to a retriever object
# Retriever provides a standard interface for document retrieval
print("🔍 Creating retriever interface...")

# search_kwargs: Configuration for the similarity search
# k: Number of documents to retrieve per query
retriever = vectorstore.as_retriever(
    search_kwargs={"k": 3}  # Return top 3 most similar chunks
)

print("✅ Retriever configured!")
print(f"   • Search type: Similarity (cosine distance)")
print(f"   • Results per query: 3 documents")
print("")

print("=" * 50)
print("💾 ChromaDB setup complete! Ready for similarity search!")
````

### 🛠️ Helper Function: Document Formatting

Before we build the chain, we need a helper function to convert retrieved Document objects into a single string that can be injected into our prompt template.

````python
# ========================================
# 🛠️ STEP 5B: DOCUMENT FORMATTING HELPER
# ========================================

print("🛠️ Creating Document Formatting Helper Function...")
print("=" * 50)

# format_docs(): Converts a list of Document objects to a single string
# This is necessary because:
# 1. The retriever returns List[Document]
# 2. The prompt template expects a string for {context}
# 3. We need to combine multiple documents with clear separation

def format_docs(docs):
    """
    Convert retrieved documents to a formatted string for prompt injection.
    
    Parameters:
    -----------
    docs : List[Document]
        List of Document objects from the retriever.
        Each Document has:
        - page_content: The text content
        - metadata: Dict with source, page, etc.
    
    Returns:
    --------
    str
        All document contents joined with double newlines.
        Format: "Doc1 content\n\nDoc2 content\n\nDoc3 content"
    
    Why this format?
    - Double newlines clearly separate documents
    - LLM can distinguish between different sources
    - Maintains readability in the prompt
    """
    # Log the formatting process
    print(f"   📄 Formatting {len(docs)} documents for context injection")
    
    # join(): Concatenates list elements with separator
    # page_content: Attribute containing the document text
    # "\n\n": Double newline separator for clear document boundaries
    formatted = "\n\n".join(doc.page_content for doc in docs)
    
    print(f"   ✅ Context formatted ({len(formatted)} characters)")
    return formatted

print("✅ format_docs() function defined!")
print("")
print("📊 Function Signature:")
print("   format_docs(docs: List[Document]) -> str")
print("")
print("🔄 Processing Flow:")
print("   [Doc1, Doc2, Doc3] → 'Doc1\\n\\nDoc2\\n\\nDoc3'")

print("=" * 50)
print("🛠️ Helper function ready!")
````

### 🔗 Building the RAG Chain with LCEL

**LCEL (LangChain Expression Language)** provides a clean, functional way to compose complex chains using the pipe `|` operator.

**Chain Composition Pattern:**
```python
chain = component1 | component2 | component3
```

Each component's output becomes the next component's input (like Unix pipes).

````python
# ========================================
# 🔗 STEP 5C: BUILD THE RAG CHAIN
# ========================================

print("🔗 Building RAG Chain with LCEL...")
print("=" * 50)

# The RAG chain connects all components:
# 1. Retriever - Finds relevant documents
# 2. Formatter - Converts docs to string
# 3. Prompt - Formats the instruction
# 4. LLM - Generates the response
# 5. Parser - Extracts the text output

print("📊 Chain Components:")
print("   1️⃣  Retriever → Find relevant documents")
print("   2️⃣  Format → Convert docs to context string")
print("   3️⃣  Prompt → Create formatted LLM input")
print("   4️⃣  LLM → Generate response (Groq)")
print("   5️⃣  Parser → Extract text from response")
print("")

# Build the chain using LCEL (pipe operator)
# Each | connects an output to the next input

# Dictionary with keys for prompt variables:
# - "context": Goes through retriever → format_docs
# - "question": Passed through unchanged
#
# RunnablePassthrough(): A special runnable that returns its input unchanged
# Used when we need to pass data through without modification

print("🔧 Assembling chain components...")

rag_chain = (
    # Step 1: Create a dictionary with context and question
    # This dict maps to our prompt template variables
    {
        # "context" key: 
        # retriever fetches docs | format_docs converts to string
        "context": retriever | format_docs,
        
        # "question" key:
        # RunnablePassthrough passes the input directly
        "question": RunnablePassthrough()
    }
    # Step 2: Format the prompt with context and question
    | prompt
    
    # Step 3: Send to LLM for generation
    | llm
    
    # Step 4: Parse the output to a string
    | StrOutputParser()
)

print("✅ RAG chain assembled!")
print("")
print("📊 Chain Architecture:")
print("""
   ┌─────────────────────────────────────────────────────────────┐
   │                      RAG CHAIN FLOW                         │
   ├─────────────────────────────────────────────────────────────┤
   │                                                             │
   │  Input: "What is RAG?"                                      │
   │           │                                                 │
   │           ├──────────────────┬───────────────────┐         │
   │           │                  │                   │         │
   │           ▼                  ▼                   │         │
   │    ┌──────────────┐   ┌─────────────┐           │         │
   │    │  Retriever   │   │ Passthrough │           │         │
   │    │ (Similarity) │   │  (Identity) │           │         │
   │    └──────────────┘   └─────────────┘           │         │
   │           │                  │                   │         │
   │           ▼                  │                   │         │
   │    ┌──────────────┐          │                   │         │
   │    │ format_docs  │          │                   │         │
   │    │  (Join)      │          │                   │         │
   │    └──────────────┘          │                   │         │
   │           │                  │                   │         │
   │           └────────┬─────────┘                   │         │
   │                    │                             │         │
   │                    ▼           {"context": ..., "question": ...}
   │             ┌──────────────┐                     │         │
   │             │   Prompt     │                     │         │
   │             │  Template    │                     │         │
   │             └──────────────┘                     │         │
   │                    │                             │         │
   │                    ▼                             │         │
   │             ┌──────────────┐                     │         │
   │             │  ChatGroq    │                     │         │
   │             │    (LLM)     │                     │         │
   │             └──────────────┘                     │         │
   │                    │                             │         │
   │                    ▼                             │         │
   │             ┌──────────────┐                     │         │
   │             │ StrOutput    │                     │         │
   │             │   Parser     │                     │         │
   │             └──────────────┘                     │         │
   │                    │                             │         │
   │                    ▼                             │         │
   │  Output: "RAG is a technique that..."           │         │
   │                                                             │
   └─────────────────────────────────────────────────────────────┘
""")

print("=" * 50)
print("🔗 RAG chain ready for queries!")
````

````python
# ========================================
# 🔁 STEP 5E: PRODUCTION-GRADE INTERACTIVE QUERY LOOP
# ========================================

def run_rag_interactive_loop():
    """
    Production-grade interactive RAG query loop.
    
    Features:
    ---------
    - Continuous querying with detailed logging
    - Query history tracking
    - Performance metrics
    - Robust error handling
    - Graceful exit handling
    
    Exit Commands: 'exit', 'quit', 'q', 'bye'
    """
    
    print("=" * 60)
    print("🚀 RAG INTERACTIVE QUERY SYSTEM")
    print("=" * 60)
    print("")
    print("📚 Knowledge Base Loaded:")
    print(f"   • Documents: {len(chunks)} chunks")
    print(f"   • Vector DB: ChromaDB (in-memory)")
    print(f"   • LLM: Groq llama-3.3-70b-versatile")
    print(f"   • Retrieval: Top 3 similar chunks")
    print("")
    print("💡 Instructions:")
    print("   • Type your question and press Enter")
    print("   • Type 'exit', 'quit', 'q', or 'bye' to end session")
    print("")
    print("-" * 60)
    
    # Initialize session tracking
    query_count = 0
    query_history = []
    total_time = 0
    session_start = time.time()
    
    # Exit commands (case-insensitive)
    exit_commands = {'exit', 'quit', 'q', 'bye', 'stop', 'end'}
    
    print("")
    print("🎯 Ready for questions! Ask me anything about your documents.")
    print("")
    
    # Main interactive loop
    # while True: Creates an infinite loop that runs until explicitly broken
    # This pattern is common for interactive applications
    while True:
        try:
            # Get user input
            # input(): Built-in function to read user input from stdin
            # Returns string (empty string if user just presses Enter)
            print("-" * 60)
            user_query = input("❓ Your Question: ").strip()
            
            # Validate input
            # strip(): Removes leading/trailing whitespace
            # lower(): Converts to lowercase for case-insensitive comparison
            
            # Check for empty input
            if not user_query:
                print("⚠️  Empty input detected. Please enter a question.")
                continue
            
            # Check for exit command
            if user_query.lower() in exit_commands:
                print("")
                print("=" * 60)
                print("👋 ENDING RAG SESSION")
                print("=" * 60)
                
                # Display session statistics
                session_duration = time.time() - session_start
                print("")
                print("📊 Session Statistics:")
                print(f"   • Total queries: {query_count}")
                if query_count > 0:
                    print(f"   • Average response time: {total_time/query_count:.2f} seconds")
                print(f"   • Session duration: {session_duration:.2f} seconds")
                print("")
                print("📜 Query History:")
                for i, q in enumerate(query_history, 1):
                    print(f"   {i}. {q[:50]}..." if len(q) > 50 else f"   {i}. {q}")
                print("")
                print("✅ Thank you for using RAG Interactive System!")
                print("=" * 60)
                break  # Exit the while loop
            
            # Process the query
            query_count += 1
            query_history.append(user_query)
            
            print("")
            print(f"📝 Processing Query #{query_count}...")
            print("-" * 40)
            
            # Step-by-step logging of RAG pipeline
            print("🔄 RAG Pipeline Execution:")
            print("")
            
            # Step 1: Retrieval
            print("   📍 Step 1/4: RETRIEVAL")
            print("   └── 🔍 Searching vector database...")
            print("   └── 🧮 Computing similarity scores...")
            
            query_start = time.time()
            
            # Step 2: Context Formation
            print("   📍 Step 2/4: CONTEXT FORMATION")
            print("   └── 📄 Retrieving top 3 relevant chunks...")
            print("   └── ✂️  Formatting context string...")
            
            # Step 3: LLM Generation
            print("   📍 Step 3/4: LLM GENERATION")
            print("   └── 📤 Sending prompt to Groq API...")
            print("   └── 🤖 Model: llama-3.3-70b-versatile")
            print("   └── ⏳ Waiting for response...")
            
            # Execute the RAG chain
            # invoke(): Runs the complete chain from input to output
            response = rag_chain.invoke(user_query)
            
            query_time = time.time() - query_start
            total_time += query_time
            
            # Step 4: Response Parsing
            print("   📍 Step 4/4: RESPONSE PARSING")
            print("   └── 📥 Response received!")
            print("   └── ✅ Output parsed successfully")
            print("")
            
            # Display the response
            print("=" * 60)
            print("🤖 AI RESPONSE:")
            print("=" * 60)
            print("")
            print(response)
            print("")
            print("-" * 60)
            print(f"⏱️  Response Time: {query_time:.2f} seconds")
            print(f"📊 Queries this session: {query_count}")
            print("-" * 60)
            print("")
            
        except KeyboardInterrupt:
            # Handle Ctrl+C gracefully
            # KeyboardInterrupt: Exception raised when user presses Ctrl+C
            print("")
            print("⚠️  Keyboard interrupt detected (Ctrl+C)")
            print("💡 Use 'exit' to quit gracefully")
            continue
            
        except Exception as e:
            # Catch-all for any other errors
            # This ensures the loop continues even if there's an error
            print("")
            print(f"❌ Error occurred: {type(e).__name__}")
            print(f"   Message: {str(e)}")
            print("💡 Please try again or type 'exit' to quit")
            print("")
            continue

# Show instructions
print("🔁 Interactive RAG Loop Defined!")
print("")
print("📋 To start the interactive session, run:")
print("   run_rag_interactive_loop()")
print("")
print("💡 The loop will:")
print("   ✓ Accept your questions continuously")
print("   ✓ Show detailed pipeline execution logs")
print("   ✓ Display response times")
print("   ✓ Track query history")
print("   ✓ Handle errors gracefully")
````
