```python
short_summary = " "
```

```python
from groq import Groq

client = Groq()


def generate_summary(s ):
    print("generating summary for ", s)
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant which can create a short summary of all the provided texts within the same context including summaries of long texts if the the text is too long. max words 30 if the text itself is < 30 keep text as it is no need to summarize it.  "
            },
            {
                "role": "user",
                "content": f" Long full texts : {s}"
            }
        ],
        model="llama-3.3-70b-versatile"
    )
    response = chat_completion.choices[0].message.content


    return response



chat_completion = client.chat.completions.create(
    messages=[
        # Set an optional system message. This sets the behavior of the
        # assistant and can be used to provide specific instructions for
        # how it should behave throughout the conversation.
        {
            "role": "system",
            "content": "You are a helpful assistant! max words 30"
        },
        # Set a user message for the assistant to respond to.
        {
            "role": "user",
            # along with query previou msgs inlcude 
            "content":  f"Context :  This is the total summary of this user including previous msgs {short_summary}  query : iran vs usa war march 2026 status"
        }
    ],

    # The language model which will generate the completion.
    model="llama-3.3-70b-versatile"
)
# print(chat_completion)
# Print the completion returned by the LLM.
response = chat_completion.choices[0].message.content

temp = response + short_summary
print("temp", temp)
short_summary = generate_summary(temp )

print( "current_response", response)
print("summary", short_summary)


```
